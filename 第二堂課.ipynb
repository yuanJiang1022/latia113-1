```python
def a(x):
  result = 0
  while result<10:
    result = result+x
    if result==5:
      return result
  return result

b = a(1)
c = a(2)

print(b)
print(c)
```

```python
import pandas as pd

dates = pd.date_range("20130101", periods = 6)
dates

df = pd.DataFrame(np.random.randn(6, 4), index = dates, columns = list("ABCD"))

print(df.loc[dates[0]])
print(df.loc[:, ["A", "B"]])
print(df.loc["20230102":"20230104", ["A", "B"]])
print(df.loc[dates[0], "A"])
```

```python
import pandas as pd

dates = pd.date_range("20130101", periods = 6)
df = pd.DataFrame(np.random.randn(6, 4), index = dates, columns = list("ABCD"))

df1 =df.reindex(index = dates[0:4], columns = list(df.columns)+["E"])
df1.loc[dates[0]:dates[1], "E"]=1
df1
```

處理

```python
import pandas as pd

dates = pd.date_range("20130101", periods = 6)
df = pd.DataFrame(np.random.randn(6, 4), index = dates, columns = list("ABCD"))

df1 =df.reindex(index = dates[0:4], columns = list(df.columns)+["E"])
df1.loc[dates[0]:dates[1], "E"]=1

print(df1.dropna())
print(df1.fillna(5))
print(df1.interpolate("backfill"))
```
