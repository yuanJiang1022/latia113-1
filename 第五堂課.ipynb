ANOVA(analysis of variance)變異數分析

```python
import numpy as np
from scipy import stats

#成績
group1 = [85, 86, 88, 75, 78, 94]
group2 = [92, 94, 89, 96, 93, 95, 76, 90]
group3 = [78, 81, 79, 84, 77, 80]

#ANOVA
f_value, p_value = stats.f_oneway(group1, group2, group3)

#輸出
print("F-value:", f_value)
print("P-value:", p_value)

#解釋
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference in scores between the groups.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in scores between the groups.")
```

應用分析

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

#創建數據集
data = {
    'score': [85, 86, 88, 75, 78, 94, 92, 94, 89, 96, 93, 95, 78, 81, 79, 84, 77, 80],
    'mathood' :['A', 'A', 'A', 'A', 'A', 'A', #A group
                'B', 'B', 'B', 'B', 'B', 'B', #B group
                'C', 'C', 'C', 'C', 'C', 'C' #C group
                ]
       }

#轉化為DataFrame
df = pd.DataFrame(data)

#ols 進行 ANOVA
model = ols('score ~ C(mathood)', data = df).fit()
anova_table = sm.stats.anova_lm(model, typ = 2)

#輸出
print(anova_table)
```

一個正式的報告

```python
import numpy as np
import pandas as pd
from scipy import stats
import statsmodels.api as sm
from statsmodels.formula.api import ols

# 設定每組數據的參數
mean_values = [375, 375, 375]  # 假設三個組別的平均值都設為375
std_dev = 50  # 假設標準差為50
size_per_group = 20  # 每組20人

# 隨機生成常態分佈的數據
group1 = np.random.normal(loc=mean_values[0], scale=std_dev, size=size_per_group)
group2 = np.random.normal(loc=mean_values[1], scale=std_dev, size=size_per_group)
group3 = np.random.normal(loc=mean_values[2], scale=std_dev, size=size_per_group)

# 裁剪數據，確保在250到500之間
group1 = np.clip(group1, 250, 500)
group2 = np.clip(group2, 250, 500)
group3 = np.clip(group3, 250, 500)

# 合併三組數據
data = np.concatenate([group1, group2, group3])

# 創建 DataFrame 方便查看結果
groups = ['Group 1'] * size_per_group + ['Group 2'] * size_per_group + ['Group 3'] * size_per_group
df = pd.DataFrame({
    'Group': groups,
    'FTP': data
})

# 顯示生成的數據
print(df.head(20))  # 顯示前20條數據

# 進行單因素 ANOVA
f_statistic, p_value = stats.f_oneway(group1, group2, group3)

# 輸出 ANOVA 分析結果
print("\nANOVA Analysis Result:")
print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")

# 解釋結果
alpha = 0.05  # 顯著性水平
if p_value < alpha:
    print("拒絕零假設：至少有一組訓練方法對 FTP 有顯著影響。")
else:
    print("接受零假設：三組訓練方法對 FTP 的影響無顯著差異。")

#分析
model = ols('Group 1', data = df).fit()
anova_table = sm.stats.anova_lm(model, typ = 2)

print(anova_table)
```
